## Matplotlib, Seaborn

## Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
##simple Examples
x = np.arange(0,10)
y = np.arange(11,21)
x
```

```python
plt.scatter(x,y)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Scatter Plot")
```

```python
y = x**3
y
```

```python
##plt plot
plt.plot(x,y,"y*",linestyle = "dashed",linewidth = 2,markersize = 12)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("2d Diagram")
```

```python
#Creating Subplots
plt.subplot(2,2,1) #(nrows,ncolumns,index)
plt.plot(x,y,"r--")
plt.subplot(2,2,2)
plt.plot(x,y,"g*--")
plt.subplot(2,2,3)
plt.plot(x,y,"bo")
plt.subplot(2,2,4)
plt.plot(x,y,"go")
```

```python
x  = np.arange(0,4*np.pi,0.1)
y = np.sin(x)
plt.plot(x,y,"y*")
```

```python
#Bar Plot
x = [2,8,10]
y = [11,16,9]

x2 = [3,9,11]
y2 = [6,15,7]

plt.bar(x,y)
plt.bar(x2,y2,color="g")

plt.title("Bar Graph")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

plt.show()
```

```python
#Histograms
a = np.array([22,87,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a)
plt.title("Histogram")
plt.show()
```

```python
#Pie Chart
#Data to Plot
labels = "Python","C++","Ruby","Java"
sizes = [215,130,245,210]
colors = ["gold","yellowgreen","lightcoral","lightskyblue"]
explode = (0.4,0,0,0) #explode 1st slice
#Plot
plt.pie(sizes,explode=explode,labels=labels,colors=colors,
       autopct="%1.1f%%",shadow=False)

plt.axis("equal")
plt.show()
```

```python
#Box Plot
data = [np.random.normal(0,std, 100) for std in range(1,4)]

#rectangular box plot
plt.boxplot(data,vert=True,patch_artist=False)
plt.show()
```

## Seaborn

```python
import seaborn as sns
df = sns.load_dataset("tips")
df.head()
```

```python
df.shape
```

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

```python
#Pair Plot
sns.pairplot(df)
plt.show()
```

```python
sns.pairplot(df,hue="time")
plt.show()
```

```python
#checking distribution of the features
sns.distplot(df["tip"])
plt.show()
```

```python
#Joint->relationship of numerical features
sns.jointplot(x="tip",y="total_bill",data=df,kind="hist")
plt.show()
```

```python
sns.jointplot(x="tip",y="total_bill",data=df,kind="hex")
plt.show()
```

```python
sns.countplot(x="sex",data=df)
plt.show()
```

```python
sns.barplot(x="sex",y="total_bill",data=df)
plt.show()
```

```python
sns.boxplot(x="day",y="total_bill",data=df)
plt.show()
```

```python
sns.boxplot("day","total_bill",hue="smoker",data=df)
plt.show()
```

```python
sns.violinplot("day","total_bill",data=df)
plt.show()
```
